# **Introdução**

   A pandemia de COVID-19, causada pelo vírus SARS-CoV-2, teve um impacto global sem precedentes, afetando a saúde pública, a economia e a sociedade como um todo (WHO, 2020). Desde a confirmação do primeiro caso em Wuhan, China, no final de 2019, a doença rapidamente se espalhou para diversos países, levando a Organização Mundial da Saúde (OMS) a declarar estado de emergência em 30 de janeiro de 2020 e pandemia em 11 de março de 2020 (PAHO, 2020). O Brasil confirmou seu primeiro caso em 26 de fevereiro de 2020, em São Paulo, o que marcou o início de uma crise sanitária de grande magnitude (FIOCRUZ, 2021).

   São Paulo, como o estado mais populoso do Brasil e um dos principais centros econômicos do país, rapidamente se tornou o epicentro da pandemia. A região apresentou os maiores índices de casos e óbitos, resultando na adoção de medidas rigorosas, como quarentenas, fechamento de comércios e restrições de mobilidade (SEADE, 2021).
    
   A análise de séries temporais é essencial para compreender a dinâmica da transmissão da doença, permitindo identificar padrões e avaliar a eficácia das medidas sanitárias implementadas ao longo do tempo. Este estudo visa analisar a evolução da COVID-19 no estado de São Paulo, utilizando dados epidemiológicos para compreender a progressão da pandemia e seus impactos.


# **Motivações e justificativa**

A pandemia de COVID-19 evidenciou a importância da análise epidemiológica na formulação de políticas públicas eficientes. Estudos indicam que o monitoramento de séries temporais permite a detecção precoce de surtos, auxilia na previsão de cenários futuros e contribui para a distribuição adequada de recursos de saúde.

No caso do estado de São Paulo, o alto volume populacional e a densidade urbana intensificaram a transmissão da doença, tornando essencial a análise do impacto das medidas restritivas, da disponibilidade de leitos hospitalares e da adesão à vacinação. Compreender essas variáveis é crucial para aprimorar futuras estratégias de combate a pandemias e otimizar a resposta do sistema de saúde pública.

# **Descrição da base de dados**

O conjunto de dados utilizado é o "Casos e Óbitos de COVID-19 no Brasil", disponibilizado pela plataforma Brasil.io. Essa base de dados reúne informações detalhadas sobre a evolução da pandemia no Brasil, consolidando registros oficiais de casos e óbitos em nível municipal e estadual.

**O dataset contém as seguintes informações principais:**

Data do registro (YYYY-MM-DD)
Município e estado (código IBGE, nome da cidade e sigla do estado)
Número de casos confirmados (diário e acumulado)
Número de óbitos (diário e acumulado)
Taxa de incidência (casos por 100 mil habitantes)
Fonte dos dados (origem da informação reportada)
Estrutura e Organização
Os dados estão estruturados em formato tabular, sendo disponibilizados em arquivos CSV e acessíveis via API. Cada linha do dataset representa um registro de um município em uma determinada data, permitindo análises temporais e geográficas.

**Os principais campos são organizados da seguinte maneira:**

Identificação geográfica: Estado, município e código IBGE
Dados epidemiológicos: Casos confirmados e óbitos
Indicadores populacionais: Taxa de incidência para contextualização dos números absolutos

**Forma e Período de Coleta:**
A coleta dos dados é realizada diariamente, com informações extraídas de boletins epidemiológicos das secretarias estaduais de saúde e outras fontes oficiais. A série histórica cobre registros desde os primeiros casos reportados no Brasil, permitindo um acompanhamento contínuo da pandemia.

**Fonte dos Dados**
Os dados são disponibilizados gratuitamente pelo Brasil.io, uma plataforma colaborativa que organiza e distribui bases públicas em formatos acessíveis para análise de dados. O acesso ao conjunto de dados pode ser feito através do link:
